```python
rho_interp_full=xr.open_dataarray('/data/pacific/iufarias/z_star/new_method_1000_100p/rho_full_M01.nc')
rho_interp_smth=xr.open_dataarray('/data/pacific/iufarias/z_star/new_method_1000_100p/rho_smth_M01.nc')
```

```python
rho_full_star=rho_interp_full.copy()
rho_smth_star=rho_interp_smth.copy()
```

```python
%%time

ix=0
rho_full_stacked=np.ravel(rho_interp_full[ix])
index_sorted_full=np.argsort(rho_full_stacked)

rho_star=rho_interp_full[ix].copy(data=rho_full_stacked[index_sorted_full].reshape(rho_interp_full[ix].shape))
# rho_star_m=rho_interp_full.copy(data=np.tile(rho_star.mean(['xi_rho','eta_rho']),
#                                              (rho_interp_full.shape[2],rho_interp_full.shape[1],1)).T)

rho_full_star[ix]=rho_star
```

```python
rho_star_reshape=rho_star.stack(profiles=['xi_rho','eta_rho'])
```

```python
mean_profile_rho_star=rho_star.mean(['xi_rho','eta_rho'])

ul_profile_rho_star=2*rho_star.std(['xi_rho','eta_rho'])+ mean_profile_rho_star
ll_profile_rho_star=-2*rho_star.std(['xi_rho','eta_rho'])+ mean_profile_rho_star



mean_profile_rho_t=rho_interp_full[ix].mean(['xi_rho','eta_rho'])
ul_profile_rho_t=2*rho_interp_full[ix].std(['xi_rho','eta_rho'])+ mean_profile_rho_t
ll_profile_rho_t=-2*rho_interp_full[ix].std(['xi_rho','eta_rho'])+ mean_profile_rho_t



mean_profile_rho=rho_interp_full.mean(['ocean_time','xi_rho','eta_rho'])
ul_profile_rho=2*rho_interp_full.std(['ocean_time','xi_rho','eta_rho'])+ mean_profile_rho
ll_profile_rho=-2*rho_interp_full.std(['ocean_time','xi_rho','eta_rho'])+ mean_profile_rho
```

```python
%%time

plt.figure(figsize=(10,8))

mean_profile_rho_star.plot(y='z_rho',color='r',label=r'$\overline{\rho_i^{*}}$')
plt.fill_betweenx(x1=ll_profile_rho_star,x2=ul_profile_rho_star,y=rho_star.z_rho,color='red',alpha=0.4)

# mean_profile_rho_t.plot(y='z_rho',color='k',label=r'$\overline{\rho}}$')
# plt.fill_betweenx(x1=ll_profile_rho_t,x2=ul_profile_rho_t,y=rho_star.z_rho,color='k',alpha=0.4)

mean_profile_rho.plot(y='z_rho',color='g',label=r'$\overline{\rho}^{\,(x,y,t)}$')
plt.fill_betweenx(x1=ll_profile_rho,x2=ul_profile_rho,y=rho_star.z_rho,color='g',alpha=0.2)



# plt.ylim(-100,0)
plt.legend()
plt.tight_layout()
```
