```python

rho_star_full=xr.open_mfdataset('/data/pacific/iufarias/tseng_z_star/rho_star/rho_star_full*.nc')['__xarray_dataarray_variable__']
rho_star_smth=xr.open_mfdataset('/data/pacific/iufarias/tseng_z_star/rho_star/rho_star_smth*.nc')['__xarray_dataarray_variable__']

# rho_star_path='/data/pacific/iufarias/tseng_z_star/rho_star/'
# rho_star_list=glob(rho_star_list +'rho_star_full*');rho_star_list.sort()
```

```python
%%time

rho_star_full_avg=[]
rho_star_smth_avg=[]

for i_t in range(rho_star_full.ocean_time.shape[0]):
# for i_t in range(30):

    rho_star_full_avg.append(rho_star_full[i_t,:21,:,:].mean(['eta_rho','xi_rho']).compute().expand_dims('ocean_time'))
    rho_star_smth_avg.append(rho_star_smth[i_t,:21,:,:].mean(['eta_rho','xi_rho']).compute().expand_dims('ocean_time'))
```

```python
rstar_full_profile=xr.concat(rho_star_full_avg,dim='ocean_time')
rstar_smth_profile=xr.concat(rho_star_smth_avg,dim='ocean_time')
```

## $\rho^*$ plots

```python
rstar_full_profile.to_netcdf('/data/pacific/iufarias/APE_global/rstar_full_profile.nc')
rstar_smth_profile.to_netcdf('/data/pacific/iufarias/APE_global/rstar_smth_profile.nc')
```

```python

mean_profile_rho_full=rstar_full_profile.mean('ocean_time')
ul_profile_rho_full=2*rstar_full_profile.std('ocean_time')+ mean_profile_rho_full
ll_profile_rho_full=-2*rstar_full_profile.std('ocean_time')+ mean_profile_rho_full



mean_profile_rho_smth=rstar_smth_profile.mean('ocean_time')
ul_profile_rho_smth=2*rstar_smth_profile.std('ocean_time')+ mean_profile_rho_smth
ll_profile_rho_smth=-2*rstar_smth_profile.std('ocean_time')+ mean_profile_rho_smth
```

```python

plt.figure(figsize=(10,8))

mean_profile_rho_full.plot(y='z_rho',color='cornflowerblue',label=r'$\overline{\rho^{*}}$ (FULL)')
plt.fill_betweenx(x1=ll_profile_rho_full,x2=ul_profile_rho_full,y=mean_profile_rho_full.z_rho,color='cornflowerblue',alpha=0.4)


mean_profile_rho_smth.plot(y='z_rho',color='tomato',label=r'$\overline{\rho^{*}}$ (SMTH)')
plt.fill_betweenx(x1=ll_profile_rho_smth,x2=ul_profile_rho_smth,y=mean_profile_rho_full.z_rho,color='tomato',alpha=0.4)

plt.xlabel('')
plt.title('')
plt.ylabel('')
# plt.ylim(-100,0)
plt.legend()
plt.tight_layout()
```

## $N^2_r$ plots

```python
dz=np.gradient(rho_star_full.z_rho).mean()
N2r_full=-(9.8/1025)*rstar_full_profile.copy(data=np.gradient(rstar_full_profile,axis=1))/dz
N2r_smth=-(9.8/1025)*rstar_smth_profile.copy(data=np.gradient(rstar_smth_profile,axis=1))/dz
```

```python
N2r_full.to_netcdf('/data/pacific/iufarias/APE_global/N2r_full_profile.nc')
N2r_smth.to_netcdf('/data/pacific/iufarias/APE_global/N2r_smth_profile.nc')
```

```python

mean_profile_N2_full=N2r_full.mean('ocean_time')
ul_profile_N2_full=2*N2r_full.std('ocean_time')+ mean_profile_N2_full
ll_profile_N2_full=-2*N2r_full.std('ocean_time')+ mean_profile_N2_full



mean_profile_N2_smth=N2r_smth.mean('ocean_time')
ul_profile_N2_smth=2*N2r_smth.std('ocean_time')+ mean_profile_N2_smth
ll_profile_N2_smth=-2*N2r_smth.std('ocean_time')+ mean_profile_N2_smth
```

```python

plt.figure(figsize=(10,8))

N2r_full.mean('ocean_time').plot(y='z_rho',color='cornflowerblue',label=r'$\overline{N^{2}_r}$ (FULL)')
plt.fill_betweenx(x1=ll_profile_N2_full,x2=ul_profile_N2_full,y=mean_profile_rho_full.z_rho,color='cornflowerblue',alpha=0.2)


N2r_smth.mean('ocean_time').plot(y='z_rho',color='tomato',label=r'$\overline{N^2_r}$ (SMTH)')
plt.fill_betweenx(x1=ll_profile_N2_smth,x2=ul_profile_N2_smth,y=mean_profile_rho_full.z_rho,color='tomato',alpha=0.2)

plt.xlabel('')
plt.title('')
plt.ylabel('')
# plt.ylim(-100,0)
plt.legend()
plt.tight_layout()
```
