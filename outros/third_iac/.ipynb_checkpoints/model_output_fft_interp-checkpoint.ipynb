```python
full_ps = xrft.power_spectrum(varf.SST[-1].drop(['nav_lat','nav_lon']),dim=['x', 'y'])
smth_ps = xrft.power_spectrum(vart.SST[-1].drop(['nav_lat','nav_lon']),dim=['x', 'y'])


fy=smth_ps.freq_y
fx=smth_ps.freq_x

fx,fy=np.meshgrid(fx,fy)



plt.figure(figsize=(23,6))
plt.subplot(1,3,1)
plt.contourf(fx,fy,np.log(full_ps),levels=np.linspace(0,24,51),cmap='rainbow')
plt.colorbar()
plt.title('FULL')

plt.subplot(1,3,2)
plt.contourf(fx,fy,np.log(smth_ps),levels=np.linspace(0,24,51),cmap='rainbow')
plt.colorbar()
plt.title('SMTH')

plt.subplot(1,3,3)
plt.contourf(fx,fy,full_ps-smth_ps,levels=np.linspace(-1000,1000,51),cmap='seismic')
plt.colorbar()
plt.title('FULL - SMTH')






xx,yy = np.meshgrid(np.arange(-varf.nav_lon.shape[1]/2,varf.nav_lon.shape[1]/2),
                    np.arange(-varf.nav_lat.shape[0]/2,varf.nav_lat.shape[0]/2))
R = np.sqrt(xx**2+yy**2)


rr=R.max()/varf.nav_lat.shape[0]
print(rr)

fPz=np.zeros(varf.nav_lat.shape[0])
tPz=np.zeros(varf.nav_lat.shape[0])
ky=np.zeros(varf.nav_lat.shape[0])

plt.figure()
plt.contourf(xlon,xlat,R)
plt.scatter(xlon[(R>=10)&(R<=100)],xlat[(R>=10)&(R<=100)],s=0.01,c='r')
plt.colorbar()




for ir in range(xlat.shape[0]-1):

    fPz[ir]=np.nansum(np.array(full_ps)[(R>=rr*ir)&(R<=(ir+1)*rr)])
    tPz[ir]=np.nansum(np.array(smth_ps)[(R>=rr*ir)&(R<=(ir+1)*rr)])
    ky[ir]=np.nansum(np.array(np.abs(fy))[(R>=rr*ir)&(R<=(ir+1)*rr)])

plt.figure(figsize=(10,8))
plt.plot(np.log(1/ky),np.log(fPz),label='FULL',linewidth=2,color='navy')
plt.plot(np.log(1/ky),np.log(tPz),label='SMTH',linewidth=2,color='lightcoral')
plt.ylabel(r'$P_{(SST)}$')
plt.xlabel(r'$\lambda$ [km]')


plt.legend()
plt.grid()

plt.figure(figsize=(10,8))
plt.plot((np.log(ky)),np.log(fPz),label='FULL',linewidth=2,color='navy')
plt.plot((np.log(ky)),np.log(tPz),label='SMTH',linewidth=2,color='lightcoral')
plt.ylabel(r'$P_{(SST)}$')
plt.xlabel(r'$K_y$ $[km^{-1}]$')


plt.legend()
plt.grid()


```
