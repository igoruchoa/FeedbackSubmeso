```python
timer=pd.DatetimeIndex(varf.time_counter.groupby("time_counter.dayofyear").mean())
```

```python
bathy=varf.h[1:-1,1:-1]
mask_bathy=np.isnan(bathy)
```

```python
fo=sw.f(bathy.nav_lat_rho)
```

```python
tau_y_full=varf.svstr.groupby("time_counter.dayofyear").mean()
tau_x_full=varf.sustr.groupby("time_counter.dayofyear").mean()


tau_y_smth=vart.svstr.groupby("time_counter.dayofyear").mean()
tau_x_smth=vart.sustr.groupby("time_counter.dayofyear").mean()
```

```python
tau_x_full,tau_y_full,_=romspickle.uvw2rho_3d(tau_x_full,tau_y_full,tau_x_full)
tau_x_full = tau_x_full.rename({'y_u': 'y_rho','x_u': 'x_rho'})
tau_y_full = tau_y_full.rename({'y_v': 'y_rho','x_v': 'x_rho'})

tau_x_smth,tau_y_smth,_=romspickle.uvw2rho_3d(tau_x_smth,tau_y_smth,tau_x_smth)
tau_x_smth = tau_x_smth.rename({'y_u': 'y_rho','x_u': 'x_rho'})
tau_y_smth = tau_y_smth.rename({'y_v': 'y_rho','x_v': 'x_rho'})
```

```python
V_ek_full=-(tau_x_full)/(1025*fo)
U_ek_full=(tau_y_full)/(1025*fo)


V_ek_smth=-(tau_x_smth)/(1025*fo)
U_ek_smth=(tau_y_smth)/(1025*fo)
```

```python
bat_thresh=1000
```

```python
V_ek_full_coast=V_ek_full.compute()
U_ek_full_coast=U_ek_full.compute()

U_ek_full_coast.data[:,bathy>=bat_thresh]=np.nan
V_ek_full_coast.data[:,bathy>=bat_thresh]=np.nan

# U_ek_full_coast.data[:,:,750:850]=np.nan
# V_ek_full_coast.data[:,:,750:850]=np.nan


V_ek_smth_coast=V_ek_smth.compute()
U_ek_smth_coast=U_ek_smth.compute()

U_ek_smth_coast.data[:,bathy>=bat_thresh]=np.nan
V_ek_smth_coast.data[:,bathy>=bat_thresh]=np.nan


# U_ek_smth_coast.data[:,:,750:850]=np.nan
# V_ek_smth_coast.data[:,:,750:850]=np.nan


```
