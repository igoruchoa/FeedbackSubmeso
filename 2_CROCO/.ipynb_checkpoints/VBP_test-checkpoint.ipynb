#### Spectral parameters

```python
dx=0.5
nf=2
wdws='hann' #'flattop','hann'
wdw_cor=True
scl='density'
```

```python
z_interp=np.linspace(-5,-50,10)
dz=np.nanmean(np.diff(z_interp))
fnum=240
```

# $\widehat{\frac{b}{N^2_r}}^* \widehat{w \frac{\partial b}{\partial z}}$

```python
N2_mean=xr.open_mfdataset('/homes/metogra/iufarias/Documents/data/new_spec/N2z_mean_full_total.nc')['__xarray_dataarray_variable__']
```

```python
bN2r_w_dbdz_spec=xr.DataArray(coords=(range(df.rho.shape[0]), z_interp,range(fnum)),
            dims=['ocean_time','z_rho','freq_r'])
# w_dbdz_spec['z_rho_psi']=N2z.z_rho_psi
```

```python
w_full_psi=df.w[:,:,120:-120,120:800]
rho_full_psi=df.rho[:,:,120:-120,120:800]
```

```python
%%time

# for iy in range(0,350):
for iz in range(df.rho.shape[0]):

    rho_interp_full=rho_full_psi[iz].xroms.isoslice(z_interp).compute()
    b_interp_full=xroms.buoyancy(rho_interp_full)
    N2_interp_full=xr.apply_ufunc(np.gradient, b_interp_full, kwargs={'axis': 0})/dz
    w_interp_full=w_full_psi[iz].xroms.isoslice(z_interp).compute()

    
    loc_sp=xrft.isotropic_cross_spectrum((b_interp_full/(N2_mean[iz].data[:,None,None])).drop(['nav_lat_rho', 'nav_lon_rho']),
                                         (w_interp_full*N2_interp_full).drop(['nav_lat_rho', 'nav_lon_rho']),
                                          dim=['eta_rho','xi_rho'],nfactor=nf,truncate='True',scaling=scl,
                                          detrend='linear',window=wdws,window_correction=wdw_cor).real

    bN2r_w_dbdz_spec[iz,:,:]=loc_sp.data
```

```python
bN2r_w_dbdz_spec.to_netcdf('/homes/metogra/iufarias/Documents/data/new_spec/bN2r_w_dbdz_total.nc')
```

### $N^2_r$

```python
rho_full=df.rho
```

```python
z=rho_full.z_rho0.compute()
```

```python
N2z=xr.DataArray(coords=(range(rho_full.shape[0]), range(z_interp.shape[0]),
                         range(df.rho[:,:,120:-120,120:800].shape[2]),
                         range(df.rho[:,:,120:-120,120:800].shape[3])),
            dims=['ocean_time','z_rho_psi','eta_rho','xi_rho'])
```

```python
%%time
dz=np.diff(z_interp).mean()
for ix in range(df.rho.shape[0]):
# for ix in range(0,350):
    # for ix in range(2):
    rho_interp_full=rho_full[:,:,120:-120,120:800][ix].xroms.isoslice(z_interp).compute()
    b_interp_full=xroms.buoyancy(rho_interp_full)
    N2_interp_full=np.gradient(b_interp_full,axis=0)/dz
    N2z[ix,:,:,:]=N2_interp_full
```

```python
%%time
N2z.to_netcdf('/homes/metogra/iufarias/Documents/data/new_spec/N2z_hor_full_total.nc')
```

### $\widehat{w N^2}$

```python
Wz=xr.DataArray(coords=(range(rho_full.shape[0]), range(z_interp.shape[0]),
                         range(df.rho[:,:,120:-120,120:800].shape[2]),
                         range(df.rho[:,:,120:-120,120:800].shape[3])),
            dims=['ocean_time','z_rho_psi','eta_rho','xi_rho'])
```

```python
%%time
w_full_psi=df.w
```

```python
%%time
# for iy in range(0,350):
for iy in range(df.rho.shape[0]):


    w_interp_full=w_full_psi[iy].xroms.isoslice(z_interp).compute()[:,120:-120,120:800]
    Wz[iy,:,:,:]=w_interp_full.data
```

```python
w_dbdz_t=N2z*Wz
w_dbdz_t['z_rho_psi']=z_interp
```

```python
w_dbdz_spec=xr.DataArray(coords=(range(df.rho.shape[0]), z_interp,range(fnum)),
            dims=['ocean_time','z_rho_psi','freq_r'])
# w_dbdz_spec['z_rho_psi']=N2z.z_rho_psi
```

```python
%%time

# for iy in range(0,350):
for iy in range(df.rho.shape[0]):

    w_dbdz_sp=xrft.isotropic_power_spectrum(w_dbdz_t[iy],
                                          dim=['eta_rho','xi_rho'],nfactor=nf,truncate='True',scaling=scl,
                                          detrend='linear',window=wdws,window_correction=wdw_cor)
    w_dbdz_spec[iy,:,:]=w_dbdz_sp.data
```

```python
w_dbdz_spec['freq_r']=w_dbdz_sp.freq_r.data
w_dbdz_spec['ocean_time']=df.ocean_time.data
w_dbdz_spec['z_rho_psi']=z_interp
```

```python
w_dbdz_spec.to_netcdf('/homes/metogra/iufarias/Documents/data/new_spec/w_dbdz_total.nc')
```
