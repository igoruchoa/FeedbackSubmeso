# Dask Array
##### Dask using "lazy" modes of computation, if we want to create a numpy array, we do this

```python
import numpy as np
shape= (1000,4000)
ones_np=np.ones(shape)
ones_np
```

```python
print('{:.1f} MB'.format(ones_np.nbytes/1e6))
```

##### Creating the same array with dask interface

```python
import dask.array as da
ones= da.ones(shape) #this is not optmized because we did not set the chunks for the array to parallel computing
ones
```

```python
shape_chunk=(1000,1000)
ones_d=da.ones(shape,chunks=shape_chunk)
ones_d
```
