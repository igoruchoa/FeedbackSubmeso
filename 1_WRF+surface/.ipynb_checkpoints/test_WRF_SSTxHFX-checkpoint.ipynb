```python
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
import gc
import xarray as xr
import pandas as pd
import xrft as xrft
from scipy.interpolate import griddata
from scipy.ndimage import gaussian_filter as gauss
from scipy.integrate import trapezoid as trapz
from pyspec import spectrum as spec
from matplotlib.colors import LogNorm
import warnings
import scipy.integrate as integ
warnings.filterwarnings("ignore")
```

```python

path='/data/pacific/lrenault/SASI/FULL/WRF/'
listm=glob(path+'*DAY.nc');listm.sort()
# varf=xr.open_dataset(listm[-9:][xmon])
varf=xr.open_mfdataset(listm[-9:])

path='/data/pacific/lrenault/SASI/SMTH/WRF/'
listt=glob(path+'*DAY.nc');listt.sort()
#vart=xr.open_dataset(listt[xmon])
vart=xr.open_mfdataset(listt)
```

```python
import datetime

t_ind=pd.DatetimeIndex(varf.time_centered.compute())
mon_ind=pd.DatetimeIndex(varf.time_centered.compute()).month
mes=np.array([11,12,1,2,3,4,5,6,7])
mes_str=['Nov/11','Dec/11','Jan/12','Feb/12','Mar/12','Apr/12','May/12','Jun/12','Jul/12']
```

```python
Tf=varf.SST[0:30,30:-30,30:200] 
Hf=varf.HFX[0:30,30:-30,30:200]


Tt=vart.SST[0:30,30:-30,30:200] 
Ht=vart.HFX[0:30,30:-30,30:200]


```

```python
exspec=xrft.isotropic_cross_spectrum(Tf[0].drop(['nav_lat','nav_lon']),Tf[0].drop(['nav_lat','nav_lon']),nfactor=1,truncate='True',detrend='linear', window=True)
Xf=np.zeros([Tf.shape[0],exspec.shape[0]])*np.nan;
Xt=np.zeros([Tt.shape[0],exspec.shape[0]])*np.nan;


for ixx in range(Tf.shape[0]):
    'Full'
    Xf_hat=xrft.isotropic_cross_spectrum(Tf[ixx].drop(['nav_lat','nav_lon']),Hf[ixx].drop(['nav_lat','nav_lon']),nfactor=1,truncate='True',detrend='linear', window=True)
    Xf_hat=Xf_hat.real

    Xf[ixx,:]=(Xf_hat.to_numpy())

    'Smth'
    Xt_hat=xrft.isotropic_cross_spectrum(Tt[ixx].drop(['nav_lat','nav_lon']),Ht[ixx].drop(['nav_lat','nav_lon']),nfactor=1,truncate='True',detrend='linear', window=True)
    Xt_hat=Xt_hat.real

    Xt[ixx,:]=(Xt_hat.to_numpy())

```

```python

XXf=np.nanmean(Xf,axis=0)
XXt=np.nanmean(Xt,axis=0)

Kk=np.array(exspec.freq_r/dxx)
Kx=(Kk[:-1]+Kk[1:])/2
dK=np.diff(Kk)[0]


XAf=integ.cumtrapz(y=XXf[::-1],x=Kk[::-1]);
XAt=integ.cumtrapz(y=XXt[::-1],x=Kk[::-1]);

XAf=-XAf[::-1]
XAt=-XAt[::-1]

fig,ax = plt.subplots(figsize=(10,4))
ax.plot(Kx,XAf,label='FULL',linewidth=2,color='purple')
ax.plot(Kx,XAt,label='SMTH',linewidth=2,color='orange')
# ax.plot([1/50,1/50],[-0.001,0.01],color='grey',linestyle='--')


ax.set_xticks(np.array([-2,-1]))
ax.set_xticklabels([r'10$^{-2}$',r'10$^{-1}$'])
ax.set_title(r'Ogive {$Co[SST\,*\,HFX]$}',fontweight='bold')
ax.set_xlabel(r'Wavenumber $ [km^{-1}]$',fontsize=15,fontweight='bold')
ax.set_ylabel(r' $E(k)$',color="k",fontsize=15,fontweight='bold')
ax.set_xlim(1e-2,1*10**-0.5)
ax.set_ylim(-0.001,0.01)
ax.set_xscale('log')



ax.legend()

# ax2=ax.twinx()
# ax2.plot(np.log10(Kx),(1-(HAf)/(HAt))*100,alpha=0.5,linestyle='dashed',color='cadetblue')
# ax2.set_ylabel(r' Ratio ($1 - \frac{FULL}{SMTH}$) [$\%$]',fontsize=15,color='cadetblue')
plt.tight_layout()

plt.savefig('/homes/metogra/iufarias/Documents/figures/Test_Ogive_cospec_SST-HFX.png')
```
