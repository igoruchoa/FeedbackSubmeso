```python
'CROCO'
path='/data/pacific/lrenault/SASI/CROCO/FULL/'
listF=glob(path+'*.nc');listF.sort()
varF=xr.open_dataset(listF[:-3][io],decode_times=False)

if io<=1:
    path='/data/pacific/lrenault/SASI/CROCO/SMTH/CROCO/'
    listT=glob(path+'*.nc');listT.sort()
    varT=xr.open_dataset(listT[io],decode_times=False)


else:
    path='/data/pacific/lrenault/SASI/CROCO/SMTH/'
    listT=glob(path+'*.nc');listT.sort()
    varT=xr.open_dataset(listT[:-3][io],decode_times=False)

```

## SST variables from CROCO

```python
TF=np.zeros([varF['temp'][::4].shape[0],varF['temp'].shape[2],varF['temp'].shape[3]])*np.nan
TT=np.zeros([varT['temp'][::4].shape[0],varT['temp'].shape[2],varT['temp'].shape[3]])*np.nan


ist=0
ien=4
i=0
for i in range(varF['temp'][::4,0,:,:].shape[0]):
    # print(i)
    # print(ist,ien)
    TF[i,:,:]=np.nanmean(varF['temp'][ist:ien,0,:,:],axis=0)
    TT[i,:,:]=np.nanmean(varT['temp'][ist:ien,0,:,:],axis=0)
    ist+=4
    ien+=4
    i+=1
```

```python
sst_F=TF
sst_T=TT
```

## Surface Heatflux from WRF

```python
# sst_F=varf['SST']

hft=vart['HFX']
hff=varf['HFX']
```

## Using griddata to compare the two resolutions for the cross spectrum

```python
xW,yW=np.arange(hft.shape[2]),np.arange(hft.shape[1])
xWW,yWW=np.meshgrid(xW,yW)

xC,yC=np.linspace(0,hft.shape[2],sst_F.shape[2]),np.linspace(0,hft.shape[1],sst_F.shape[1])
xCC,yCC=np.meshgrid(xC,yC)

hfT=np.zeros(sst_T.shape)*np.nan;
hfF=np.zeros(sst_F.shape)*np.nan;

for i in range(hft.shape[0]):
    
    hfT[i,:,:]=griddata((np.ravel(xWW),np.ravel(yWW)),np.ravel(hft[i]),(xCC,yCC),method='linear');
    hfF[i,:,:]=griddata((np.ravel(xWW),np.ravel(yWW)),np.ravel(hff[i]),(xCC,yCC),method='linear');
```

## Cropping Areas

```python
Hf=xr.DataArray(hfF)[:,100:-100,90:700]
Ht=xr.DataArray(hfT)[:,100:-100,90:700]


Tf=xr.DataArray(sst_F)[:,100:-100,90:700]
Tt=xr.DataArray(sst_T)[:,100:-100,90:700]
```

```python
exspec=xrft.isotropic_cross_spectrum(Tf[0],Hf[0],nfactor=1,truncate='True',detrend='linear', window=True)
CRf=np.zeros([Tf.shape[0],exspec.shape[0]])*np.nan;
CRt=np.zeros([Tf.shape[0],exspec.shape[0]])*np.nan;


for ixx in range(Hf.shape[0]):
    'Full'
    Cf_co=xrft.isotropic_cross_spectrum(Tf[ixx],Hf[ixx],nfactor=1,truncate='True',detrend='linear', window=True)
    Cf_co=Cf_co.real

    'Smth'
    Ct_co=xrft.isotropic_cross_spectrum(Tt[ixx],Ht[ixx],nfactor=1,truncate='True',detrend='linear', window=True)
    Ct_co=Ct_co.real


    CRf[ixx,:]=Cf_co
    CRt[ixx,:]=Ct_co

```

```python
plt.figure()
plt.plot(exspec.freq_r/dxx,Cf_co)
# plt.yscale('symlog',linthreshy=1e-5)

plt.figure()
plt.plot(exspec.freq_r/dxx,Cf_co.real)
# plt.yscale('symlog',linthreshy=1e-5)
```

```python
tf=sst_F[:,100:-100,90:700]
tt=sst_T[:,100:-100,90:700]

exspec=xrft.isotropic_powerspectrum(tf[0],nfactor=1,truncate='True',detrend='linear', window=True)

Tf=np.zeros([tf.shape[0],exspec.shape[0]])*np.nan;
Tt=np.zeros([tt.shape[0],exspec.shape[0]])*np.nan;

for ixx in range(tf.shape[0]):
    tpf=xrft.isotropic_power_spectrum(tf[ixx],nfactor=1,truncate='True',detrend='linear', window=True)
    tpt=xrft.isotropic_power_spectrum(tt[ixx],nfactor=1,truncate='True',detrend='linear', window=True)

    Tf[ixx,:]=(tpf.to_numpy())
    Tt[ixx,:]=(tpt.to_numpy())
```
